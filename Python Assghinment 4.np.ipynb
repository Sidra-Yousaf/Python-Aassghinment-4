```python
# Q NO.4 : Write a Python program to sum all the numeric items in a dictionary.
l=[1,2,3,4,5,2,3,4,7,9,5]
l1=[]
for i in l:
    if i not in l1:
        l1.append(i)
    else:
        print(i,end=' ')
```

```python
# Q no.6: Make a calculator using Python with addition , subtraction ,
# multiplication ,division and power.
#Python program to perform Addition, Subtraction,
#Multiplication and division of two numbers
num1=int(input("Enter the first number: "))
num2=int(input("Enter the second number: "))
print("Enter the operator you want to perform");
ch=input("Enter any of these operator for operation +, -, *, /  ")
result=0
if ch=='+':
    result=num1+num2;
elif ch=='-':
    result=num1-num2;
elif ch=='*':
    result=num1*num2;
elif ch=='/':
    result=num1/num2;
else:
   print("char is not supported");
print(num1,ch,num2,": ",result)
```
